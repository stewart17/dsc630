```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data
dodgers = pd.read_csv("dodgers-2022.csv")
```

```python
# look at data
dodgers
```

```python
# histogram of attendance
sns.histplot(x = "attend", data= dodgers)
plt.title("Attendance Histogram")
plt.xlabel("Attendance")
plt.show()
```

```python
# temperature influence on attendance
sns.catplot(x="temp", y = "attend", data=dodgers, kind="bar", aspect=2)
plt.title("Temperature Relationship to Attendance")
plt.ylabel("Attendance")
plt.xlabel("Temperature")
plt.show()
```

```python
# month and attendance
sns.catplot(x="month", y = "attend", data=dodgers, kind="point", aspect=2)
plt.title("Time of Year Relationship to Attendance")
plt.ylabel("Attendance")
plt.xlabel("Month")
plt.show()
```

```python
# month and attendance with bobblehead in consideration
sns.catplot(x="month", y = "attend", data=dodgers, kind="bar", aspect=2, hue= "bobblehead")
plt.title("Bobblehead Relationship to Attendance by Month")
plt.ylabel("Attendance")
plt.xlabel("Month")
plt.show()
```

```python
# month and attend barplot with shirt
sns.catplot(x="month", y = "attend", data=dodgers, kind="bar", aspect=2, hue= "shirt")
plt.title("Shirt Relationship to Attendance by Month")
plt.ylabel("Attendance")
plt.xlabel("Month")
plt.show()
```

```python
# cap influence on attendance
sns.catplot(x="month", y = "attend", data=dodgers, kind="bar", aspect=2, hue= "cap")
plt.title("Caps Relationship to Attendance by Month")
plt.ylabel("Attendance")
plt.xlabel("Month")
plt.show()
```

```python
# fireworks influence on attendance
sns.catplot(x="month", y = "attend", data=dodgers, kind="bar", aspect=2, hue= "fireworks")
plt.title("Fireworks Relationship to Attendance by Month")
plt.ylabel("Attendance")
plt.xlabel("Month")
plt.show()
```

## Recommendations

The analysis above demonstrates clear markers of what drives attendance. We can broadly classify those markers as things that can be changed and those that cannot (or cannot *easily* be changed in the case of changing what months games are held in). A good strategy to drive attendance would be to use the things that can be changed to drive attendance in moments where the factors that cannot be changed are causing a drop in attendance.

To arrive at this conclusion I first took a histogram of attendance to see the distribution of attendance. It looks to be normally distributed with a maximum attendance of around 50,000 people. I then used bar graphs and line plots to compare the different relationships between attendance and these factors. Once I saw that attendance saw a wide variation according to month. I compared the other factors according to the month, with the goal to see if there was a relationship. With all that in mind, I then thought about categorizing these factors as follows:

The factors that cannot be changed which influence attendance are the temperature and the months the games are played during. As can be seen in the bar plot comparing temperature to attendance, attendance is highest when temperatures are in the 70s and then drop off on the high and low end of the spectrum. Likewise, attendance markedly rises from May into June then steadily drops off as the games move into fall.

The factors that can be changed which influence attendance are bobbleheads and shirts. Fireworks and caps had no difference on attendance. Having bobbleheads and shirts available saw an increase in attendance during those periods. In the case of shirts, it saw a rise in 5000 people and for bobbleheads a staggering 10,000 person difference in some cases. 

My recommendation towards increasing attendance is to more consistently have bobbleheads and shirts throughout the season. The hopes is that by having this more readily available it encourages people to come to the games during the less busy periods. Bobbleheads are already being sold throughout the season mostly, and the influence can clearly be seen. Yet despite a huge effect on attendance, shirts are only being sold during two months. 

In addition, during the months where attendance is naturally higherâ€”June, July, and Augustâ€”there should be a concerted effort made towards having raffles to win shirts and bobbleheads. The opportunity to possibly win some of this merchandise without fully investing in the full cost may help to drive attendance.

A final recommendation is to experiment with other forms of merchandise to measure their influence on attendance. For instance, perhaps in the colder months selling more sweatshirts and beanies could help drive more people to the games.
